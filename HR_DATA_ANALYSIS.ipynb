```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('HR Data.csv')
```

```python
data.head()
```

```python
data.columns
```

 Removing unnecessary columns.

```python
columns_to_drop = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours']
data.drop(columns=columns_to_drop, inplace=True)
```

```python
data.columns
```

Giving the columns new names

```python
data.rename(columns={'BusinessTravel':'Travel Frequency','EnvironmentSatisfaction':'Satisfaction'}, inplace=True)
```

sanitizing specific columns.

```python
# Impute missing values with mean for numerical columns
data['Age'].fillna(data['Age'].mean(), inplace=True)
```

```python
# Impute missing values with mode for categorical columns
data['Travel Frequency'].fillna(data['Travel Frequency'].mode()[0], inplace=True)
```

```python
# Correcting data types
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
```

```python
# Standardizing formats
data['Travel Frequency'] = data['Travel Frequency'].str.strip().str.lower()
```

Removing null values and duplicates

```python
data.dropna(inplace=True)
data.drop_duplicates()
```

Balancing the Dataset

```python
from sklearn.utils import resample

# Separate majority and minority classes
data_majority = data[data.Attrition == 'No']

data_minority = data[data.Attrition == 'Yes']

# Upsample minority class
data_minority_upsampled = resample(data_minority,
                                 replace=True,     # sample with replacement
                                 n_samples=len(data_majority),    # to match majority class
                                 random_state=123) # reproducible results

# Combine majority class with upsampled minority class
data_balanced = pd.concat([data_majority, data_minority_upsampled])

# Display new class counts
data_balanced.Attrition.value_counts()
```
